<p style="background-color:#blue;color:black;font-size:20px;text-align:center;border-radius:10px 10px;font-weight:bold;border:2px solid black;"> The Monte Hall Problem 🏆 <span style='font-size:28px; background-color:blue ;'>  </span></p>

```python
# setting up the door numbers
doors = [1, 2, 3]

# simulating the placement of the prize
prize = np.random.choice(doors, 1)
print(prize)

# simulating the choice of the contestant
contestant_choice = np.random.choice(doors, 1)
print(contestant_choice)

# determining the doors which can be opened by the assistant
doors_that_can_be_opened_by_assistant = np.delete(doors, [prize-1, contestant_choice-1])
print(doors_that_can_be_opened_by_assistant)

# simulating the choice of the assistant
door_opened_by_assistant = np.random.choice(doors_that_can_be_opened_by_assistant, 1)
print(door_opened_by_assistant)

# determining the doors which can be opened by the contestant if he chooses to switch
doors_that_can_be_opened_by_contestant = np.delete(doors, [contestant_choice-1, door_opened_by_assistant-1])
print(doors_that_can_be_opened_by_contestant)

# simulating the choice of the contestant who is switching
switch_door = np.random.choice(doors_that_can_be_opened_by_contestant, 1)
print(switch_door)

# determining if the contestant would win by switching or would win by staying
if switch_door == prize:
    print("Wins by switching")
else:
    print("Loses by switching")
        
```

```python
# approximation of the probability of winning by switching
switch_wins = 0
total_sim_num = 100000
doors = list(range(1,4))
for i in range(total_sim_num):
    prize = np.random.choice(doors, 1)
    contestant_choice = np.random.choice(doors,1) 
    doors_that_can_be_opened_by_assistant = np.delete(doors,[prize-1,contestant_choice-1]) 
    door_opened_by_assistant = np.random.choice(doors_that_can_be_opened_by_assistant,1)
    doors_that_can_be_opened_by_contestant = np.delete(doors,[contestant_choice-1,door_opened_by_assistant-1]) 
    switch_door = np.random.choice(doors_that_can_be_opened_by_contestant,1) 
    if switch_door == prize:
        switch_wins = switch_wins + 1

sim_prob_switch_and_win = float(switch_wins)/float(total_sim_num)
print(sim_prob_switch_and_win)
```

#### Let's adapt the Monte Hall simulation program for: n = 3, 4, 5, 10, 20, 50, 100.

```python
# approximation of the probability of winning by switching
prob = []
for k in [3, 4, 5, 10, 20, 50, 100]:
    switch_wins = 0
    total_sim_num = 100000
    doors = list(range(1,k+1))
    for i in range(total_sim_num):
        prize = np.random.choice(doors, 1)
        contestant_choice = np.random.choice(doors,1) 
        doors_that_can_be_opened_by_assistant = np.delete(doors,[prize-1,contestant_choice-1]) 
        door_opened_by_assistant = np.random.choice(doors_that_can_be_opened_by_assistant,1)
        doors_that_can_be_opened_by_contestant = np.delete(doors,[contestant_choice-1,door_opened_by_assistant-1]) 
        switch_door = np.random.choice(doors_that_can_be_opened_by_contestant,1) 
        if switch_door == prize:
            switch_wins = switch_wins + 1
    sim_prob_switch_and_win = float(switch_wins)/float(total_sim_num)
    prob.append(sim_prob_switch_and_win)
print("The probabilities of winning by switching for each of the values of n are respectively:", prob)
```
